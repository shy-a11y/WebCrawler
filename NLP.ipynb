```python
#1페이지에 있는 모든 리포트 들에 대해서 제목과 본문을 komoran으로 분석한 다음, 형태소의 종류별(명사갯수, 형용사 갯수, 동사 갯수, ...) 갯수를 구하고, 이것을 DataFrame에 저장하고
#excel파일로 추출한다.

url = "You need to put your own url"
html = urlopen(url)
bs_obj = BeautifulSoup(html, "html.parser")

data4=[]
komoran = Komoran()
thispage = bs_obj.find_all("td",{"class":"on"})
thispage_a_href= thispage[0].find("a")['href']
print(thispage_a_href)
pagenumber = int(thispage_a_href[-1])
print(pagenumber)
nextpage=0
for i in range(3):
  nextpage= thispage_a_href[:-1]+str(pagenumber)
  url="You need to put your own url"+nextpage
  print("현재 서치하고 있는 페이지 url: " + url)
  pagenumber+=1
  #pagenumber가 변하면서 url이 변하고, url이 변하면 bs_obj가 변한다
  html = urlopen(url)
  bs_obj = BeautifulSoup(html, "html.parser")
  count = len(bs_obj.find_all('td', {"style":"padding-left:10"}))
  print("한 페이지에서 종목 수:"+str(count))
  #한 페이지 내부에서 종목명, 날짜, 제목, 본문 뽑아내는 코드
  for k in range(count):
    #종목명 추출
    company_name_td_all = bs_obj.find_all('td', {"style":"padding-left:10"})
    company_name = company_name_td_all[i].find('a',{"class":"stock_item"})
    #날짜 추출
    report_date = bs_obj.find_all('td', {"class":"date","style":"padding-left:5px"})
    #종목명 옆 제목 추출
    table = bs_obj.find("table",{"class":"type_1"})
    tableData_1 = table.find_all("td",{"style":"padding-left:10"})
    report_title = tableData_1[k].next_sibling.next_sibling
    print("title of the report is "+report_title.text)
    #본문 내용 추출
    report_page=report_title.find("a")['href']
    report_url = "You need to put your own url" +report_page
    print("report itself url is "+report_url)
    html2= urlopen(report_url)
    bs_obj2 = BeautifulSoup(html2,"html.parser")
    report_table = bs_obj2.find("table",{"summary":"종목분석 리포트 본문내용"})
    report_table_2=report_table.find("td",{"class":"view_cnt"})
    every_p_tag = report_table_2.find_all("p")
    count2 = len(report_table_2.find_all('p'))
    report_itself=""
    for t in range(count2):
      report_itself += every_p_tag[t].text
    print("report itself is "+report_itself)

    # Komoran을 통한 형태소 분석
    title_morphs = komoran.pos(report_title.text)
    content_morphs = komoran.pos(report_itself)
    #형태소 갯수 카운
    morph_count = {
            "명사": sum(1 for word, pos in title_morphs + content_morphs if pos == "NNG" or pos == "NNP"),
            "형용사": sum(1 for word, pos in title_morphs + content_morphs if pos == "VA"),
            "동사": sum(1 for word, pos in title_morphs + content_morphs if pos == "VV"),
            "부사": sum(1 for word, pos in title_morphs + content_morphs if pos == "MAG"),
            "관형사": sum(1 for word, pos in title_morphs + content_morphs if pos == "MM"),
            # 필요 시 추가 형태소 (예: 감탄사, 조사 등) 갯수
        }
     # 데이터를 리스트에 추가
    data4.append([company_name.text, report_date[i].text, report_title, report_itself, morph_count["명사"], morph_count["형용사"], morph_count["동사"],
                      morph_count["부사"], morph_count["관형사"]])


df4= pd.DataFrame(data4, columns=["종목명","날짜","제목","본문","명사 갯수","형용사 갯수","동사 갯수","부사 갯수","관형사 갯수"])
print(df4)

#엑셀로 저장
excel_file_path = 'Your own path'
df4.to_excel(excel_file_path, index=False, engine='openpyxl')  # 또는 engine='xlsxwriter'

```
